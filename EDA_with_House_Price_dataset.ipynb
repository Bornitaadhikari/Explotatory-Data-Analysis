```python
# Box plot
plt.figure(figsize=(8, 6))
plt.boxplot(df["bedrooms"])
plt.xlabel("Bedrooms")
plt.ylabel("Number of Bedrooms")
plt.title("Distribution of Bedrooms using Box Plot")
plt.grid(True)
plt.show()
```

```python
# Bubble chart
plt.figure(figsize=(10, 6))
plt.scatter(df["bedrooms"], df["bathrooms"], s=df["price"] / 1000, alpha=0.7, c="blue", edgecolors="black")
plt.xlabel("Number of Bedrooms")
plt.ylabel("Number of Bathrooms")
plt.title("Bubble Chart: Bedrooms, Bathrooms, and Price")
plt.grid(True)
plt.show()
```
